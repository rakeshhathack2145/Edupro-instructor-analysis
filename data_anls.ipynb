```python
# ============================================
# NOTEBOOK 2: EXPLORATORY DATA ANALYSIS
# ============================================

# %% [markdown]
# # EduPro Analysis - Exploratory Data Analysis
# 
# Objective: Understand data distributions and relationships

# %% Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# %% Load Cleaned Data
teachers = pd.read_csv('../data/teachers_clean.csv')
courses = pd.read_csv('../data/courses_clean.csv')
transactions = pd.read_csv('../data/transactions_clean.csv')

print("✅ Data loaded successfully!")
print(f"Teachers: {teachers.shape}")
print(f"Courses: {courses.shape}")
print(f"Transactions: {transactions.shape}")

# %% Data Integration
print("\n" + "=" * 60)
print("DATA INTEGRATION")
print("=" * 60)

# Merge transactions with courses
df = transactions.merge(courses, on='CourseID', how='left')
print(f"\n✅ Step 1: Merged transactions + courses: {df.shape}")

# Merge with teachers
df = df.merge(teachers, on='TeacherID', how='left')
print(f"✅ Step 2: Merged with teachers: {df.shape}")

# Display integrated data
print("\n📊 Integrated Dataset Preview:")
print(df.head())

# Save integrated data
df.to_csv('../data/integrated_data.csv', index=False)
print("\n✅ Integrated data saved!")

# %% Rating Distributions
print("\n" + "=" * 60)
print("RATING DISTRIBUTIONS")
print("=" * 60)

fig, axes = plt.subplots(2, 2, figsize=(15, 10))

# Teacher Rating Distribution
axes[0, 0].hist(df['TeacherRating'], bins=30, color='skyblue', edgecolor='black')
axes[0, 0].axvline(df['TeacherRating'].mean(), color='red', 
                    linestyle='--', label=f'Mean: {df["TeacherRating"].mean():.2f}')
axes[0, 0].set_title('Teacher Rating Distribution', fontsize=14, fontweight='bold')
axes[0, 0].set_xlabel('Rating')
axes[0, 0].set_ylabel('Frequency')
axes[0, 0].legend()
axes[0, 0].grid(alpha=0.3)

# Course Rating Distribution
axes[0, 1].hist(df['CourseRating'], bins=30, color='lightgreen', edgecolor='black')
axes[0, 1].axvline(df['CourseRating'].mean(), color='red', 
                    linestyle='--', label=f'Mean: {df["CourseRating"].mean():.2f}')
axes[0, 1].set_title('Course Rating Distribution', fontsize=14, fontweight='bold')
axes[0, 1].set_xlabel('Rating')
axes[0, 1].set_ylabel('Frequency')
axes[0, 1].legend()
axes[0, 1].grid(alpha=0.3)

# Years of Experience Distribution
axes[1, 0].hist(df['YearsOfExperience'], bins=25, color='salmon', edgecolor='black')
axes[1, 0].axvline(df['YearsOfExperience'].mean(), color='darkred', 
                    linestyle='--', label=f'Mean: {df["YearsOfExperience"].mean():.2f}')
axes[1, 0].set_title('Years of Experience Distribution', fontsize=14, fontweight='bold')
axes[1, 0].set_xlabel('Years')
axes[1, 0].set_ylabel('Frequency')
axes[1, 0].legend()
axes[1, 0].grid(alpha=0.3)

# Age Distribution
axes[1, 1].hist(df['Age'], bins=25, color='plum', edgecolor='black')
axes[1, 1].axvline(df['Age'].mean(), color='purple', 
                    linestyle='--', label=f'Mean: {df["Age"].mean():.2f}')
axes[1, 1].set_title('Instructor Age Distribution', fontsize=14, fontweight='bold')
axes[1, 1].set_xlabel('Age')
axes[1, 1].set_ylabel('Frequency')
axes[1, 1].legend()
axes[1, 1].grid(alpha=0.3)

plt.tight_layout()
plt.savefig('../outputs/figures/distributions.png', dpi=300, bbox_inches='tight')
plt.show()

print("\n✅ Distribution plots created!")

# %% Box Plots for Ratings
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Teacher Rating by Gender
df.boxplot(column='TeacherRating', by='Gender', ax=axes[0])
axes[0].set_title('Teacher Rating by Gender', fontsize=14, fontweight='bold')
axes[0].set_xlabel('Gender')
axes[0].set_ylabel('Rating')

# Course Rating by Level
df.boxplot(column='CourseRating', by='CourseLevel', ax=axes[1])
axes[1].set_title('Course Rating by Level', fontsize=14, fontweight='bold')
axes[1].set_xlabel('Course Level')
axes[1].set_ylabel('Rating')
plt.suptitle('')  # Remove default title
plt.tight_layout()
plt.savefig('../outputs/figures/boxplots.png', dpi=300, bbox_inches='tight')
plt.show()

# %% Categorical Analysis
print("\n" + "=" * 60)
print("CATEGORICAL ANALYSIS")
print("=" * 60)

fig, axes = plt.subplots(2, 2, figsize=(16, 12))

# Expertise Distribution
expertise_counts = df['Expertise'].value_counts()
axes[0, 0].barh(expertise_counts.index, expertise_counts.values, color='teal')
axes[0, 0].set_title('Instructors by Expertise', fontsize=14, fontweight='bold')
axes[0, 0].set_xlabel('Count')
axes[0, 0].grid(axis='x', alpha=0.3)

# Course Category Distribution
category_counts = df['CourseCategory'].value_counts()
axes[0, 1].barh(category_counts.index, category_counts.values, color='coral')
axes[0, 1].set_title('Enrollments by Course Category', fontsize=14, fontweight='bold')
axes[0, 1].set_xlabel('Count')
axes[0, 1].grid(axis='x', alpha=0.3)

# Experience Category Distribution
exp_counts = df['ExperienceCategory'].value_counts()
axes[1, 0].bar(exp_counts.index, exp_counts.values, color='gold')
axes[1, 0].set_title('Instructors by Experience Level', fontsize=14, fontweight='bold')
axes[1, 0].set_ylabel('Count')
axes[1, 0].tick_params(axis='x', rotation=45)
axes[1, 0].grid(axis='y', alpha=0.3)

# Course Level Distribution
level_counts = df['CourseLevel'].value_counts()
axes[1, 1].bar(level_counts.index, level_counts.values, color='mediumpurple')
axes[1, 1].set_title('Enrollments by Course Level', fontsize=14, fontweight='bold')
axes[1, 1].set_ylabel('Count')
axes[1, 1].grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.savefig('../outputs/figures/categorical_analysis.png', dpi=300, bbox_inches='tight')
plt.show()

# %% Correlation Analysis
print("\n" + "=" * 60)
print("CORRELATION ANALYSIS")
print("=" * 60)

# Select numerical columns
numerical_cols = ['Age', 'YearsOfExperience', 'TeacherRating', 'CourseRating']
correlation_matrix = df[numerical_cols].corr()

print("\n📊 Correlation Matrix:")
print(correlation_matrix)

# Heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', 
            center=0, square=True, linewidths=1,
            cbar_kws={"shrink": 0.8}, fmt='.3f')
plt.title('Correlation Matrix - Numerical Variables', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.savefig('../outputs/figures/correlation_matrix.png', dpi=300, bbox_inches='tight')
plt.show()

# %% Key Statistics Summary
print("\n" + "=" * 60)
print("KEY STATISTICS SUMMARY")
print("=" * 60)

summary_stats = {
    'Metric': [
        'Total Enrollments',
        'Unique Instructors',
        'Unique Courses',
        'Avg Teacher Rating',
        'Avg Course Rating',
        'Avg Years Experience',
        'Avg Instructor Age'
    ],
    'Value': [
        len(df),
        df['TeacherID'].nunique(),
        df['CourseID'].nunique(),
        df['TeacherRating'].mean(),
        df['CourseRating'].mean(),
        df['YearsOfExperience'].mean(),
        df['Age'].mean()
    ]
}

summary_df = pd.DataFrame(summary_stats)
summary_df['Value'] = summary_df['Value'].round(2)

print("\n", summary_df.to_string(index=False))

# Save summary
summary_df.to_csv('../outputs/tables/summary_statistics.csv', index=False)

# %% [markdown]
# ## ✅ Key Findings from EDA
# 
# 1. Rating Distributions: Both teacher and course ratings show...
# 2. Experience Impact: Initial correlation observed...
# 3. Category Insights: Most popular categories are...
# 4. Quality Patterns: Higher experience correlates with...

print("\n" + "=" * 60)
print("✅ NOTEBOOK 2 COMPLETED SUCCESSFULLY!")
print("=" * 60)
```
