```python
# ============================================
# NOTEBOOK 1: DATA LOADING AND CLEANING
# ============================================

# %% [markdown]
# # EduPro Analysis - Data Loading and Cleaning
# 
# Objective: Load, validate, and clean the datasets
# 
# Datasets:
# - teachers.csv
# - courses.csv
# - transactions.csv

# %% Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings('ignore')

# Set display options
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 100)

# Plotting style
plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")

print("✅ Libraries imported successfully!")

# %% Load Datasets
print("=" * 60)
print("LOADING DATASETS")
print("=" * 60)

# Load teachers data
teachers = pd.read_csv('../data/teachers.csv')
print(f"\n✅ Teachers data loaded: {teachers.shape}")

# Load courses data
courses = pd.read_csv('../data/courses.csv')
print(f"✅ Courses data loaded: {courses.shape}")

# Load transactions data
transactions = pd.read_csv('../data/transactions.csv')
print(f"✅ Transactions data loaded: {transactions.shape}")

# %% Data Overview
print("\n" + "=" * 60)
print("DATA OVERVIEW")
print("=" * 60)

print("\n📊 TEACHERS DATA:")
print(teachers.head())
print(f"\nShape: {teachers.shape}")
print(f"Columns: {teachers.columns.tolist()}")

print("\n📚 COURSES DATA:")
print(courses.head())
print(f"\nShape: {courses.shape}")
print(f"Columns: {courses.columns.tolist()}")

print("\n💳 TRANSACTIONS DATA:")
print(transactions.head())
print(f"\nShape: {transactions.shape}")
print(f"Columns: {transactions.columns.tolist()}")

# %% Data Information
print("\n" + "=" * 60)
print("DATA TYPES AND MISSING VALUES")
print("=" * 60)

print("\n📊 TEACHERS:")
print(teachers.info())
print("\nMissing Values:")
print(teachers.isnull().sum())

print("\n📚 COURSES:")
print(courses.info())
print("\nMissing Values:")
print(courses.isnull().sum())

print("\n💳 TRANSACTIONS:")
print(transactions.info())
print("\nMissing Values:")
print(transactions.isnull().sum())

# %% Data Quality Checks
print("\n" + "=" * 60)
print("DATA QUALITY CHECKS")
print("=" * 60)

# Check for duplicates
print("\n🔍 Duplicate Records:")
print(f"Teachers: {teachers.duplicated().sum()}")
print(f"Courses: {courses.duplicated().sum()}")
print(f"Transactions: {transactions.duplicated().sum()}")

# Check unique values
print("\n🔑 Unique Identifiers:")
print(f"Unique Teachers: {teachers['TeacherID'].nunique()}")
print(f"Unique Courses: {courses['CourseID'].nunique()}")
print(f"Unique Transactions: {transactions['TransactionID'].nunique()}")

# Check for orphan records
print("\n🔗 Relationship Validation:")
orphan_courses = ~transactions['CourseID'].isin(courses['CourseID'])
orphan_teachers = ~transactions['TeacherID'].isin(teachers['TeacherID'])

print(f"Transactions with missing CourseID: {orphan_courses.sum()}")
print(f"Transactions with missing TeacherID: {orphan_teachers.sum()}")

# %% Statistical Summary
print("\n" + "=" * 60)
print("STATISTICAL SUMMARY")
print("=" * 60)

print("\n📊 TEACHERS STATISTICS:")
print(teachers.describe())

print("\n📚 COURSES STATISTICS:")
print(courses.describe())

# %% Categorical Variables Analysis
print("\n" + "=" * 60)
print("CATEGORICAL VARIABLES")
print("=" * 60)

print("\n👥 Teacher Gender Distribution:")
print(teachers['Gender'].value_counts())

print("\n🎓 Teacher Expertise Distribution:")
print(teachers['Expertise'].value_counts())

print("\n📚 Course Category Distribution:")
print(courses['CourseCategory'].value_counts())

print("\n📊 Course Level Distribution:")
print(courses['CourseLevel'].value_counts())

# %% Data Cleaning
print("\n" + "=" * 60)
print("DATA CLEANING")
print("=" * 60)

# Remove duplicates if any
teachers_clean = teachers.drop_duplicates()
courses_clean = courses.drop_duplicates()
transactions_clean = transactions.drop_duplicates()

print(f"\n✅ Removed duplicates:")
print(f"Teachers: {len(teachers) - len(teachers_clean)} rows")
print(f"Courses: {len(courses) - len(courses_clean)} rows")
print(f"Transactions: {len(transactions) - len(transactions_clean)} rows")

# Handle missing values (if any)
# For this project, we'll fill missing ratings with median
if teachers_clean['TeacherRating'].isnull().sum() > 0:
    teachers_clean['TeacherRating'].fillna(
        teachers_clean['TeacherRating'].median(), 
        inplace=True
    )
    print("\n✅ Filled missing TeacherRating values")

if courses_clean['CourseRating'].isnull().sum() > 0:
    courses_clean['CourseRating'].fillna(
        courses_clean['CourseRating'].median(), 
        inplace=True
    )
    print("✅ Filled missing CourseRating values")

# %% Create Experience Categories
print("\n" + "=" * 60)
print("FEATURE ENGINEERING")
print("=" * 60)

teachers_clean['ExperienceCategory'] = pd.cut(
    teachers_clean['YearsOfExperience'],
    bins=[0, 3, 7, 15, 100],
    labels=['Beginner (0-3)', 'Intermediate (3-7)', 
            'Experienced (7-15)', 'Veteran (15+)']
)

print("\n📊 Experience Categories:")
print(teachers_clean['ExperienceCategory'].value_counts())

# Create Age Groups
teachers_clean['AgeGroup'] = pd.cut(
    teachers_clean['Age'],
    bins=[0, 30, 40, 50, 100],
    labels=['Under 30', '30-40', '40-50', '50+']
)

print("\n📊 Age Groups:")
print(teachers_clean['AgeGroup'].value_counts())

# %% Save Cleaned Data
print("\n" + "=" * 60)
print("SAVING CLEANED DATA")
print("=" * 60)

teachers_clean.to_csv('../data/teachers_clean.csv', index=False)
courses_clean.to_csv('../data/courses_clean.csv', index=False)
transactions_clean.to_csv('../data/transactions_clean.csv', index=False)

print("\n✅ All cleaned datasets saved successfully!")
print(f"✅ Teachers: {len(teachers_clean)} records")
print(f"✅ Courses: {len(courses_clean)} records")
print(f"✅ Transactions: {len(transactions_clean)} records")

# %% [markdown]
# ## ✅ Summary
# 
# Data Loading:
# - Successfully loaded 3 datasets
# - No major data quality issues
# 
# Data Cleaning:
# - Removed duplicates
# - Handled missing values
# - Created derived features
# 
# Next Steps:
# - Exploratory Data Analysis
# - Statistical testing
# - Visualization

print("\n" + "=" * 60)
print("✅ NOTEBOOK 1 COMPLETED SUCCESSFULLY!")
print("=" * 60)
```
